$ x_{n+1} = \frac{1}{2} \cdot \left( x_n + \frac{c}{x_n} \right) $ converges to $\sqrt{c}$

```python
def squareRoot(c):
    a = c
    b = 0.5 * (a + (c / a) )
    step = 1

    while abs(a-b) > 0.00001:
        a = b
        b = 0.5 * (a + c / a)
        step += 1

    print(step)
    return b
```

```python
x = 3
sqrtX = squareRoot(x)

print('SquareRoot of {} is {} error is: {}'.format(x, sqrtX, abs(x - sqrtX * sqrtX)) )
```
